```python
import pandas as pd
import glob
import os
import gc

# INPUT parquet files (your Drive folder)
input_files = sorted(glob.glob(
    "/content/drive/MyDrive/Colab Notebooks/LegalAIDatasets/1. textdata/3.1 processed_chunks/*.parquet"
))

# OUTPUT folder (cleaned parquet files)
output_dir = "/content/drive/MyDrive/Colab Notebooks/LegalAIDatasets/1. textdata/3.2_clean_parquet/"

os.makedirs(output_dir, exist_ok=True)

start_index = 1


# ---------------- CLEAN FUNCTION ----------------
def clean_series(s):
    s = s.astype(str)
    s = s.str.replace("\n", " ", regex=False)
    s = s.str.replace("\r", " ", regex=False)
    s = s.str.replace(r"\s+", " ", regex=True)
    s = s.str.replace(r"^—\s*", "", regex=True)
    s = s.str.replace("§", "Section ")
    return s.str.strip()


# ---------------- MAIN LOOP ----------------
for i, file in enumerate(input_files[start_index:], start=start_index):

    try:
        df = pd.read_parquet(file)

        if "text" not in df.columns:
            continue

        # ---------------- BUILD NEW 4 COLUMNS ----------------
        df = pd.DataFrame({
            "id": "pq_" + str(i) + "_" + df.index.astype(str),
            "text": df["text"].astype(str),
            "source": "parquet",
            "media_type": "text"
        })

        # CLEAN TEXT
        df["text"] = clean_series(df["text"])

        # REMOVE EMPTY TEXT
        df = df[df["text"].str.len() > 0]

        # ---------------- SAVE AS PARQUET ----------------
        output_path = os.path.join(output_dir, f"cleaned_{i}.parquet")

        df.to_parquet(
            output_path,
            index=False
        )

        print(f"✅ Saved cleaned parquet chunk {i} | rows: {len(df)}")

        del df
        gc.collect()

    except Exception as e:
        print(f"❌ Error at chunk {i}: {e}")
        break


print("🏁 ALL CLEAN PARQUET FILES CREATED")
```
